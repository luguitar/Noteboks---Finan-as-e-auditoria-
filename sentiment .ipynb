```python
# Define a function to classify each candle
def classify_candle(open, high, low, close):
    if open > close:
        body_color = 'filled'
    else:
        body_color = 'hollow'

body_height = abs(close - open)
    wick_height = high - max(open, close)
    shadow_height = body_height - wick_height
    #  Process hollow candles -> Close > Open
    if body_height > 2 * wick_height and body_color == 'hollow':
        sentiment = 'bullish'
    elif body_height > wick_height and body_color == 'hollow':
        sentiment = 'bullish'
    elif wick_height > body_height and body_color == 'hollow':
        sentiment = 'bearish'
    elif wick_height < body_height / 2 and body_color == 'hollow':
        sentiment = 'very bullish'
    elif shadow_height < body_height / 2 and body_color == 'hollow':
        sentiment = 'bullish'
    elif shadow_height > body_height / 2 and body_color == 'hollow':
        sentiment = 'bullish'
    #  Process filled candles -> Open > Close
    elif body_height > 2 * wick_height and body_color == 'filled':
        sentiment = 'bearish'
    elif body_height > wick_height and body_color == 'filled':
        sentiment = 'bearish'
    elif wick_height > body_height and body_color == 'filled':
        sentiment = 'bullish'
    elif wick_height < body_height / 2 and body_color == 'filled':
        sentiment = 'bearish'
    elif shadow_height < body_height / 2 and body_color == 'filled':
        sentiment = 'very bearish'
    elif shadow_height > body_height / 2 and body_color == 'filled':
        sentiment = 'bearish'
    else:
        sentiment = 'neutral'
    return sentiment
```

```python
# Long
long_entry_ind = 0
long_exit_ind = 0
if self.ema_50[-1] > self.ema_200[-1]:
    long_entry_ind += 1
#  Evaluate candle sentiment
sentiment = classify_candle(self.data.Open[-1], self.data.High[-1], self.data.Low[-1], self.data.Close[-1])
if 'bullish' in sentiment == 'bullish' and 'bullish' in self.last_sentiment:
    long_entry_ind += 1
#  Determine exit
if 'bearish' in sentiment and 'bearish' in self.last_sentiment:
    long_exit_ind = 1
#  Store the last sentiment
self.last_sentiment = sentiment
#  Perform trades
if self.position.size == 0 and long_entry_ind >= 2:
    self.buy()
if self.position.size > 0 and long_exit_ind == 1:
    self.position.close()
```
