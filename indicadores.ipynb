```python
def get_current_stock_price(symbol):
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    url = f'https://financialmodelingprep.com/api/v3/quote/{symbol}?apikey={api_key}'
    response = requests.get(url)
    data = response.json()
    return data[0]['price']  # Assuming the first item contains the current price

nvda_stock_price = get_current_stock_price('NVDA')
print(f"Current NVDA Stock Price: {nvda_stock_price}")
```

```python
def get_eps(symbol):
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    url = f'https://financialmodelingprep.com/api/v3/income-statement/{symbol}?apikey={api_key}&limit=1'  # Fetching the most recent one
    response = requests.get(url)
    data = response.json()
    return data[0]['eps']

nvda_eps = get_eps('NVDA')
print(f"NVDA's EPS: {nvda_eps}")
```

```python
def get_total_shares_outstanding(symbol):
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    url = f'https://financialmodelingprep.com/api/v3/profile/{symbol}?apikey={api_key}'
    response = requests.get(url)
    data = response.json()
    return data[0]['mktCap'] / data[0]['price']  # Calculate shares outstanding

def get_book_value_per_share(symbol):
    total_shares_outstanding = get_total_shares_outstanding(symbol)
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    balance_sheet_url = f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{symbol}?apikey={api_key}&limit=1'
    response = requests.get(balance_sheet_url)
    balance_sheet_data = response.json()

    book_value = balance_sheet_data[0]['totalAssets'] - balance_sheet_data[0]['totalLiabilities']
    book_value_per_share = book_value / total_shares_outstanding
    return book_value_per_share

# Example usage for NVIDIA (NVDA)
nvda_book_value_per_share = get_book_value_per_share('NVDA')
print(f"NVIDIA's Book Value per Share: {nvda_book_value_per_share}")
```

```python
def get_dividend_per_share(symbol):
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    url = f'https://financialmodelingprep.com/api/v3/profile/{symbol}?apikey={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        dividend_per_share = data[0].get('lastDiv', 0)
        return dividend_per_share
    else:
        print(f"Error fetching data: HTTP {response.status_code}")
        return 0

nvda_dividend_per_share = get_dividend_per_share('NVDA')
print(f"NVIDIA's Dividend Per Share: {nvda_dividend_per_share}")
```

```python
def calculate_pe_ratio(stock_price, earnings_per_share):
    pe_ratio = stock_price / earnings_per_share
    return pe_ratio

Nvidia_pe_ratio = calculate_pe_ratio(nvda_stock_price, nvda_eps)
print(f"Nvidia's P/E Ratio: {Nvidia_pe_ratio}")
```

```python
def calculate_pb_ratio(stock_price, book_value_per_share):
    pb_ratio = stock_price / book_value_per_share
    return pb_ratio

Nvidia_pb_ratio = calculate_pb_ratio(nvda_stock_price, nvda_book_value_per_share)
print(f"Nvidia's P/B Ratio: {Nvidia_pb_ratio}")
```

```python
def calculate_dividend_yield(dividend_per_share, stock_price):
    dividend_yield = dividend_per_share / stock_price
    return dividend_yield

nvda_dividend_yield = calculate_dividend_yield(nvda_dividend_per_share, nvda_stock_price)
print(f"Nvidia's Dividend Yield: {nvda_dividend_yield}")
```

```python
def get_tickers_by_sector(sector, exchange, market_cap_minimum, number_of_companies, api_key):
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    url = f'https://financialmodelingprep.com/api/v3/stock-screener?marketCapMoreThan={market_cap_minimum}&volumeMoreThan=10000&sector={sector}&exchange={exchange}&limit={number_of_companies}&apikey={api_key}'
    response = requests.get(url)
    data = response.json()
    tickers = [company['symbol'] for company in data]
    return tickers

# Example usage
sector = 'Technology'
exchange = 'NASDAQ'
market_cap_minimum = '1000000000'  # 1 billion
number_of_companies = 100  # Number of companies to fetch
tickers = get_tickers_by_sector(sector, exchange, market_cap_minimum, number_of_companies, api_key)
print(tickers)
```

```python
# Function to get financial data for a ticker
def get_financial_data(ticker, api_key):
    api_key = '680c64465fdbbc5b65b3fe116ac89d18'
    income_statement_url = f'https://financialmodelingprep.com/api/v3/income-statement/{ticker}?apikey={api_key}&limit=1'
    balance_sheet_url = f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{ticker}?apikey={api_key}&limit=1'
    cash_flow_url = f'https://financialmodelingprep.com/api/v3/cash-flow-statement/{ticker}?apikey={api_key}&limit=1'

    income_statement = requests.get(income_statement_url).json()
    balance_sheet = requests.get(balance_sheet_url).json()
    cash_flow = requests.get(cash_flow_url).json()

    return income_statement[0], balance_sheet[0], cash_flow[0]

financial_data = {}
for ticker in tickers:
    try:
        income_statement, balance_sheet, cash_flow = get_financial_data(ticker, api_key)
        financial_data[ticker] = {
            'Income Statement': income_statement,
            'Balance Sheet': balance_sheet,
            'Cash Flow': cash_flow
        }
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
```

```python
stock_prices = {}

for ticker in tickers:
  stock_prices[ticker] = get_current_stock_price(ticker)
```

```python
def calculate_ratios(financial_data, stock_prices):
    ratios = {}
    for ticker, data in financial_data.items():
        try:
            stock_price = stock_prices[ticker]  # Assuming stock price is available in stock_prices dictionary

            # P/E Ratio = Stock Price / EPS
            if data['Income Statement']['epsdiluted'] == 0:
              continue  # Skip this ticker
            else:
              pe_ratio = stock_price / data['Income Statement']['epsdiluted']

            # P/B Ratio = Stock Price / (Book Value / Total Shares Outstanding)
            total_shares = data['Balance Sheet']['commonStock'] / stock_price  # Assuming common stock value represents total equity at stock price

            if total_shares == 0:
              continue  # Skip this ticker
            else:
              book_value_per_share = (data['Balance Sheet']['totalAssets'] - data['Balance Sheet']['totalLiabilities']) / total_shares
              pb_ratio = stock_price / book_value_per_share

            # Dividend Yield = Dividends Paid / (Stock Price * Total Shares)
              dividend_yield = data['Cash Flow']['dividendsPaid'] / (stock_price * total_shares)

              ratios[ticker] = {'P/E Ratio': pe_ratio, 'P/B Ratio': pb_ratio, 'Dividend Yield': dividend_yield}
        except KeyError as e:
            print(f"Data missing for {ticker}: {e}")

    return ratios

# Calculate ratios for all companies
company_ratios = calculate_ratios(financial_data, stock_prices)
```

```python
import numpy as np

def calculate_industry_averages(company_ratios):
    pe_ratios = []
    pb_ratios = []
    dividend_yields = []

    for ratios in company_ratios.values():
        if 'P/E Ratio' in ratios:
            pe_ratios.append(ratios['P/E Ratio'])
        if 'P/B Ratio' in ratios:
            pb_ratios.append(ratios['P/B Ratio'])
        if 'Dividend Yield' in ratios:
            dividend_yields.append(ratios['Dividend Yield'])

    industry_averages = {
        'P/E Ratio': np.mean(pe_ratios) if pe_ratios else None,
        'P/B Ratio': np.mean(pb_ratios) if pb_ratios else None,
        'Dividend Yield': np.mean(dividend_yields) if dividend_yields else None
    }

    return industry_averages

# Calculate industry averages
industry_averages = calculate_industry_averages(company_ratios)
```

```python
def is_stock_undervalued(ticker, stock_price, company_ratios, industry_averages):
    if ticker in company_ratios:
        ratios = company_ratios[ticker]
        pe_undervalued = ratios['P/E Ratio'] < industry_averages['Average P/E Ratio'] if 'P/E Ratio' in ratios and 'Average P/E Ratio' in industry_averages else False
        pb_undervalued = ratios['P/B Ratio'] < industry_averages['Average P/B Ratio'] if 'P/B Ratio' in ratios and 'Average P/B Ratio' in industry_averages else False

        if pe_undervalued or pb_undervalued:
            return True
        else:
            return False
    else:
        return "Ticker not found in dataset"

stock_name = "NVDA"

result = is_stock_undervalued(stock_name, get_current_stock_price(stock_name), company_ratios, industry_averages)
print(f"Is {stock_name} undervalued? {result}")
```
