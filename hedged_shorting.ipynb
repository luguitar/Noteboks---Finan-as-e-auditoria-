# Shorting strategy with a call option hedge

This notebook shows the possibility of hedging when shorting with a call option. While without hedging, the shorting strategy would theoretically have an unlimited risk, with the use of a call option, the risk of maximum possible loss is limited. 

```python
import matplotlib.pyplot as pyplot
import numpy as np
import pandas as pd
```

```python
#Actual price is 40, and I am shorting. 
#Price of the call option that I am using for the hedge on strike price 50 = 4
#Fee for the short and buy of the option is fixed at 0.3  
```

```python
lenght=120         #Lenght of x_axis on chart

initial_price=40   #Price of the shorted equtity on t(o)
call_strike=50     #Strike price of the call option
call_price=10      #Price of the call option
fixed_costs=0.3    #Fixed costs related to all transactions

expiration_price=np.linspace(0,lenght,lenght+1)   #Possible various price of an equity when expires
```

```python
pnl=initial_price-expiration_price-call_price-fixed_costs  #Calculates PnL
```

```python
df = pd.DataFrame({
        'actual':expiration_price,
        'pnl':pnl
        })
```

```python
#When i>call_strike, then the call option is ITM and will be executed

for i in range(0,lenght+1):
    if i>call_strike:
        df['pnl'][i]=initial_price-call_strike-call_price-fixed_costs
```

```python
df['pnl'].plot(figsize=(12,4),label="PnL")
pyplot.xlabel("Price of an underlying asset")
pyplot.ylabel("Profit")
pyplot.legend()
pyplot.axhline(0,color="black",linestyle="--")
pyplot.title(f"Payoff function for the equity shorting strategy hedged with one call option")
pyplot.show()
```
